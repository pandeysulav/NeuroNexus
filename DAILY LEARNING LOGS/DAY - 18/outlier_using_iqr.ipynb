```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(16,5))

plt.subplot(1,2,1)
sns.histplot(df['cgpa'], kde=True, stat='density',edgecolor='none')

plt.subplot(1,2,2)
sns.histplot(df['placement_exam_marks'], kde=True, stat='density',edgecolor='none')

plt.show()
```

```python
df['placement_exam_marks'].describe()
```

```python
sns.boxplot(x=df['placement_exam_marks'])
```

```python
# Finding the IQR
percentile25 = df['placement_exam_marks'].quantile(0.25)
percentile75 = df['placement_exam_marks'].quantile(0.75)
```

```python
percentile75
```

```python
iqr = percentile75 - percentile25
```

```python
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python

print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

Finding Outliers

```python
df[df['placement_exam_marks'] > upper_limit]
```

```python
df[df['placement_exam_marks'] < lower_limit]
```

Trimming

```python
new_df = df[df['placement_exam_marks'] < upper_limit]
```

```python
new_df.shape
```

```python
plt.figure(figsize=(16,8))

# Original data - Histogram
plt.subplot(2,2,1)
sns.histplot(df['placement_exam_marks'],
             kde=True,
             stat='density',
             edgecolor='none')

# Original data - Boxplot (horizontal)
plt.subplot(2,2,2)
sns.boxplot(x=df['placement_exam_marks'])

# New data - Histogram
plt.subplot(2,2,3)
sns.histplot(new_df['placement_exam_marks'],
             kde=True,
             stat='density',
             edgecolor='none')

# New data - Boxplot (horizontal)
plt.subplot(2,2,4)
sns.boxplot(x=new_df['placement_exam_marks'])

plt.show()
```

Capping

```python
new_df_cap = df.copy()

new_df_cap['placement_exam_marks'] = np.where(
    new_df_cap['placement_exam_marks'] > upper_limit,
    upper_limit,
    np.where(
        new_df_cap['placement_exam_marks'] < lower_limit,
        lower_limit,
        new_df_cap['placement_exam_marks']
    )
)
```

```python
new_df_cap.shape
```

```python
# Comparing

plt.figure(figsize=(16,8))

plt.subplot(2,2,1)
sns.histplot(df['placement_exam_marks'],
             kde=True,
             stat='density',
             edgecolor='none')

plt.subplot(2,2,2)
sns.boxplot(x=df['placement_exam_marks'])

plt.subplot(2,2,3)
sns.histplot(new_df_cap['placement_exam_marks'],
             kde=True,
             stat='density',
             edgecolor='none')

plt.subplot(2,2,4)
sns.boxplot(x=new_df_cap['placement_exam_marks'])

plt.show()
```
