```python
#Specifying columns data types

data = ('a,b,c,d \n'
        '1,2,3,4 \n'
        '5,6,7,8\n'
        '9,10,11')
```

```python
print(data)
```

```python
df = pd.read_csv(StringIO(data),dtype=float)
```

```python
df
```

```python
df['a'][1]
```

```python
df = pd.read_csv(StringIO(data), dtype = {'b':int,'c':np.float,'a': 'Int64'})
```

READ JSON TO CSV

```python
import pandas as pd
import json

Data = [
    {"Name": "Alice", "Age": 25, "City": "Kathmandu"},
    {"Name": "Bob", "Age": 30, "City": "Pokhara"},
    {"Name": "Charlie", "Age": 28, "City": "Lalitpur"}
]



json_str = json.dumps(Data)   # Convert list of dicts to JSON string
df = pd.read_json(json_str)
print(df)
```
