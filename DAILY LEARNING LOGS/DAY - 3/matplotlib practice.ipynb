MATPLOTLIB


```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#Simple Examples

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
#plotting using matplotlib


#plt scatter

plt.scatter(x,y, c='b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.savefig("Try.png")
```

```python
y = x*x
```

```python
plt.plot(x,y,"r*", linestyle = 'dashed',linewidth = 2, markersize= 12)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('2d diagram')
```

```python
#Subplots

plt.subplot(2,2,1)
plt.plot(x, y, 'r--')
plt.subplot(2,2,2)
plt.plot(x, y, 'b*-')
plt.subplot(2,2,3)
plt.plot(x, y, 'g*')
```

```python
x = np.arange(0,4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave representation in Matplotlib")

plt.plot(x,y)
plt.show()
```
