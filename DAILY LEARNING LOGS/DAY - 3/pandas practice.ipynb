READ JSON TO CSV

```python
import pandas as pd
import json

Data = [
    {"Name": "Alice", "Age": 25, "City": "Kathmandu"},
    {"Name": "Bob", "Age": 30, "City": "Pokhara"},
    {"Name": "Charlie", "Age": 28, "City": "Lalitpur"}
]



json_str = json.dumps(Data)   # Convert list of dicts to JSON string
df = pd.read_json(json_str)
print(df)
```

```python
df.to_json(orient="records")
```
