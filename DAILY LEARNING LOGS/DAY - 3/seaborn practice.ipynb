SEABORN TUTORIAL

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

CORRELATION

```python
df.corr(numeric_only=True)
```

```python
sns.heatmap(df.corr(numeric_only=True))
```

JOINPLOT


```python
sns.jointplot(x = 'tip',y = 'total_bill', data = df, kind='hex')
```

```python
sns.jointplot(x = 'tip',y = 'total_bill', data = df, kind='reg')
```

PAIRPLOT

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
df['smoker'].value_counts()
```

DIST PLOT

```python
sns.displot(df['tip'],kde = True)
```

```python
sns.displot(df['tip'], kde=False,bins = 10 ) #kde = kernel density estimation
```

CATEGORICAL PLOTS

```python
import seaborn as sns

sns.countplot(x='sex', data=df)  # Count on x-axis
```

```python
print("doing well")
```

```python
sns.countplot(x = 'day',data=df)
```

```python
sns.barplot(y='total_bill',x = 'sex', data = df)
```

BOX PLOT

```python
sns.boxplot(x = 'smoker', y = 'total_bill', data = df, palette='rainbow')
```

```python
sns.boxplot(data = df, orient='v')
```

```python
sns.boxplot(x= "total_bill", y = "day", hue = "smoker", data = df)
```

VIOLIN PLOT

```python
sns.violinplot(x="total_bill", y = "day", data = df, palette='rainbow')
```
