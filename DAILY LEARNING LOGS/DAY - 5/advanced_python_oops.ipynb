```python
class Car:
    def __init__(self, window, door, enginetype):
        self.windows = window
        self.doors = door
        self.enginetype = enginetype
    def self_driving(self):
        return "This is a {} car".format(self.enginetype)
```

```python
car1 = Car(4,5,"petrol")
car1.self_driving()
```

```python
car2 = Car(3,4,"diesel")
car2.self_driving()
```

```python
print(car1.windows)
```

```python
print(car2.enginetype)
```

```python
car2 = Car()
```

```python
car2.windows = 2
car2.doors = 2
```

```python
car2.enginetype = 'petrol'
```

```python
print(car2.enginetype)
```
