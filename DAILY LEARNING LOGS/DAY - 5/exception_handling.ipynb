```python
#try else code example
try:
    #code block where exception can occur 
    a = int(input("Enter first number:\n"))
    b = int(input("Enter second number:\n"))
    c = a + b
    d = a*b
    e = a/b

except TypeError:
    print("Try to make the datatype similar")
except NameError:
    print("The user have not defined in the variable.")
except ZeroDivisionError:
    print("Please give b > 0 because division by zero is not permitted.")
except Exception as ex:
    print(ex)

else:
    print(f"The addition of numbers is : {c}")
    print(f"The product of numbers is : {d}")
    print(f"The division of numbers is : {e}")
```

```python
# try else finally code structure 
#try else code example
try:
    #code block where exception can occur 
    a = int(input("Enter first number:\n"))
    b = int(input("Enter second number:\n"))
    c = a + b
    d = a*b
    e = a/b

except TypeError:
    print("Try to make the datatype similar")
except NameError:
    print("The user have not defined in the variable.")
except ZeroDivisionError:
    print("Please give b > 0 because division by zero is not permitted.")
except Exception as ex:
    print(ex)

else:
    print(f"The addition of numbers is : {c}")
    print(f"The product of numbers is : {d}")
    print(f"The division of numbers is : {e}")
finally:
    print("The execution is executed successfully.")
```
