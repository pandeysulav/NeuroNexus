```python
import asyncio
from playwright.async_api import async_playwright
import nest_asyncio

nest_asyncio.apply()

async def get_html():
    async with async_playwright() as p:
        browser = await p.chromium.launch()
        page = await browser.new_page()
        await page.goto('https://www.ambitionbox.com/list-of-companies?page=1')
        await page.wait_for_load_state('networkidle')
        html = await page.content()
        await browser.close()
        return html

webpage = asyncio.run(get_html())
print(webpage)
```
