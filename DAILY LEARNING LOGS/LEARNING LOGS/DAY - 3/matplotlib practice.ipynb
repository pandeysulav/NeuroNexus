```python
#Simple Examples

x = np.arange(0,10)
y = np.arange(11,21)
```

```python
#plotting using matplotlib


#plt scatter

plt.scatter(x,y, c='b')
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Graph in 2D")
plt.savefig("Try.png")
```

```python
y = x*x
```

```python
plt.plot(x,y,"r*", linestyle = 'dashed',linewidth = 2, markersize= 12)
plt.xlabel("X axis")
plt.ylabel('Y axis')
plt.title('2d diagram')
```

```python
#Subplots

plt.subplot(2,2,1)
plt.plot(x, y, 'r--')
plt.subplot(2,2,2)
plt.plot(x, y, 'b*-')
plt.subplot(2,2,3)
plt.plot(x, y, 'g*')
```

```python
np.pi
```

```python
x = np.arange(0,4*np.pi, 0.1)
y = np.sin(x)
plt.title("Sine Wave representation in Matplotlib")

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0, 5*np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

plt.subplot(2,1,1)
plt.plot(x, y_sin, "g^")
plt.title("Sine Graph")

plt.subplot(2,1,2)
plt.plot(x, y_cos, "ro")   # you accidentally plotted y_sin earlier
plt.title("Cosine Graph")

plt.tight_layout()   # <-- fixes spacing
plt.show()
```

```python
x = [2,3,4]
y = [5,6,7]

x2 = [16,17,19]
y2 = [3,4,7]
plt.bar(x, y, color = 'b')
plt.bar(x2, y2, color='g')

plt.title('Bar Graph')
plt.ylabel('Y Axis')
plt.xlabel('X Axis')

plt.show()
```

HISTOGRAMS

```python
a = np.array([22,67,75,43,34,56,47,73,55,54,20,51,79,31,27])
plt.hist(a,bins=20)
plt.title('histogram')
plt.show()
```

```python
data = [np.random.normal(0, std, 100) for std in range(1, 4)]

# Rectangular box plot
plt.boxplot(data, vert=True,patch_artist=True)

plt.title("Box Plot")
plt.xlabel("Samples")
plt.ylabel("Values")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['yellowgreen', 'lightcoral', 'lightskyblue', 'gold']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(
    sizes,
    explode=explode,
    labels=labels,
    colors=colors,
    autopct='%1.1f%%',
    shadow=True
)

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title("Programming Language Popularity")
plt.show()
```
